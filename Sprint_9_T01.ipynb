### TokenizaciÃ³n

Una vez que el texto se encuentra en minÃºsculas, procedemos a realizar la tokenizaciÃ³n, que consiste en dividir el texto en palabras individuales.

```python
from nltk.tokenize import word_tokenize

tokens = nltk.word_tokenize(text)
tokens
```

DespuÃ©s de realizar la tokenizaciÃ³n, hemos observado que los signos de puntuaciÃ³n se estÃ¡n considerando como palabras individuales. Para solucionar esto, procedemos a eliminar los signos de puntuaciÃ³n del texto tokenizado.

**Limpieza**

```python
tokens_cleaned = [word for word in tokens if word.isalpha()]
tokens_cleaned
```

```python
from nltk import FreqDist

freq_dist = FreqDist(tokens_cleaned)

df_tokens = pd.DataFrame([(word, freq) for word, freq in freq_dist.items()], columns=['Palabra', 'Frecuencia'])
df_tokens.sort_values(by='Frecuencia', ascending=False).reset_index(drop=True)
```

Una vez eliminados los signos de puntuaciÃ³n, hemos aplicado el mÃ©todo `FreqDist` para determinar la frecuencia de uso de cada palabra y hemos creado un dataframe para una mejor visualizaciÃ³n de los resultados.

El dataset contiene un total de 352 palabras. Al analizar los datos, se observa que las cinco palabras mÃ¡s frecuentes son las siguientes: "mr" (36 ocurrencias), "the" (36 ocurrencias), "to" (34 ocurrencias), "and" (30 ocurrencias) y "darcy" (29 ocurrencias).

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color='white')
wordcloud.generate_from_frequencies(freq_dist)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

La visualizaciÃ³n de la nube de palabras (wordcloud) confirma de manera mÃ¡s visual lo que observamos en la tabla anterior.

### Elimina las stopwords y realiza el stemming en tu conjunto de datos.

**Stopwords**

Ahora aplicaremos el filtro de stopwords a nuestro conjunto de palabras. Las stopwords son palabras comunes en inglÃ©s que suelen tener un bajo valor semÃ¡ntico y se eliminan para enfocarnos en las palabras mÃ¡s relevantes del texto.

```python
from nltk.corpus import stopwords
filtered_tokens_stopwords = [token for token in tokens_cleaned if token not in stopwords.words('english')]
filtered_tokens_stopwords
```

```python
freq_dist_stopwords = FreqDist(filtered_tokens_stopwords)

df_tokens_stopwords = pd.DataFrame([(word, freq) for word, freq in freq_dist_stopwords.items()], 
                                   columns=['Palabra', 'Frecuencia'])
df_tokens_stopwords.sort_values(by='Frecuencia', ascending=False).reset_index(drop=True)
```

Tras eliminar las stopwords, lo primero que notamos es que el nÃºmero de palabras se redujo a 300, lo que implica que habÃ­a 52 palabras sin un valor semÃ¡ntico relevante.

Debido a la eliminaciÃ³n de estas palabras, la lista de las palabras mÃ¡s frecuentes se ve modificada a partir del segundo lugar. Ahora, las palabras mÃ¡s comunes son "darcy" (29 ocurrencias), "elizabeth" (25 ocurrencias), "bingley" (13 ocurrencias) y "bennet" (8 ocurrencias).

```python
wordcloud = WordCloud(width=800, height=400, background_color='white')
wordcloud.generate_from_text(' '.join(filtered_tokens_stopwords))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

La visualizaciÃ³n de la nube de palabras (wordcloud) confirma de manera mÃ¡s impactante lo que hemos observado en la tabla anterior. Sin embargo, resulta interesante notar que la palabra "mr" aparece dos veces en nuestra representaciÃ³n visual.

**Stemming**

A continuaciÃ³n, empleamos el mÃ©todo de stemming, el cual se refiere a la reducciÃ³n de palabras a su raÃ­z o forma base. Esto implica eliminar sufijos y prefijos para obtener la parte central o significativa de una palabra.

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in filtered_tokens_stopwords]
stemmed_tokens
```

```python
freq_dist_stemmed = FreqDist(stemmed_tokens)

df_stemmed = pd.DataFrame([(word, freq) for word, freq in freq_dist_stemmed.items()], columns=['Palabra', 'Frecuencia'])

df_stemmed.sort_values(by='Frecuencia', ascending=False).reset_index(drop=True)
```

Al observar el proceso de eliminaciÃ³n de stopwords y aplicaciÃ³n de stemming, notamos que el nÃºmero de palabras se redujo de 300 a 277. Esta reducciÃ³n se debe a que, al aplicar el proceso de stemming, algunas palabras fueron reducidas a su forma base, lo que implica una pÃ©rdida parcial de su significado original. El stemming se basa en reglas heurÃ­sticas para eliminar sufijos y prefijos, sin considerar el contexto o el significado completo de las palabras.

Una observaciÃ³n interesante es que la palabra "mr" ahora tiene 40 ocurrencias. Recordemos que en la visualizaciÃ³n de la nube de palabras con stopwords, notamos repeticiones de la palabra "mr". Al aplicar el stemming, estas repeticiones se han unificado en una sola forma, lo que explica el incremento en el recuento de ocurrencias de "mr".

```python
wordcloud = WordCloud(width=800, height=400, background_color='white')
wordcloud.generate_from_text(' '.join(stemmed_tokens))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
