### Tokenización

Una vez que el texto se encuentra en minúsculas, procedemos a realizar la tokenización, que consiste en dividir el texto en palabras individuales.

```python
from nltk.tokenize import word_tokenize

tokens = nltk.word_tokenize(text)
tokens
```

Después de realizar la tokenización, hemos observado que los signos de puntuación se están considerando como palabras individuales. Para solucionar esto, procedemos a eliminar los signos de puntuación del texto tokenizado.

**Limpieza**

```python
tokens_cleaned = [word for word in tokens if word.isalpha()]
tokens_cleaned
```

```python
from nltk import FreqDist

freq_dist = FreqDist(tokens_cleaned)

df_tokens = pd.DataFrame([(word, freq) for word, freq in freq_dist.items()], columns=['Palabra', 'Frecuencia'])
df_tokens.sort_values(by='Frecuencia', ascending=False).reset_index(drop=True)
```

Una vez eliminados los signos de puntuación, hemos aplicado el método `FreqDist` para determinar la frecuencia de uso de cada palabra y hemos creado un dataframe para una mejor visualización de los resultados.

El dataset contiene un total de 352 palabras. Al analizar los datos, se observa que las cinco palabras más frecuentes son las siguientes: "mr" (36 ocurrencias), "the" (36 ocurrencias), "to" (34 ocurrencias), "and" (30 ocurrencias) y "darcy" (29 ocurrencias).

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color='white')
wordcloud.generate_from_frequencies(freq_dist)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

La visualización de la nube de palabras (wordcloud) confirma de manera más visual lo que observamos en la tabla anterior.

## Ejercicio 2

### Elimina las stopwords y realiza el stemming en tu conjunto de datos.

**Stopwords**

Ahora aplicaremos el filtro de stopwords a nuestro conjunto de palabras. Las stopwords son palabras comunes en inglés que suelen tener un bajo valor semántico y se eliminan para enfocarnos en las palabras más relevantes del texto.

```python
from nltk.corpus import stopwords
filtered_tokens_stopwords = [token for token in tokens_cleaned if token not in stopwords.words('english')]
filtered_tokens_stopwords
```

```python
freq_dist_stopwords = FreqDist(filtered_tokens_stopwords)

df_tokens_stopwords = pd.DataFrame([(word, freq) for word, freq in freq_dist_stopwords.items()], 
                                   columns=['Palabra', 'Frecuencia'])
df_tokens_stopwords.sort_values(by='Frecuencia', ascending=False).reset_index(drop=True)
```

Tras eliminar las stopwords, lo primero que notamos es que el número de palabras se redujo a 300, lo que implica que había 52 palabras sin un valor semántico relevante.

Debido a la eliminación de estas palabras, la lista de las palabras más frecuentes se ve modificada a partir del segundo lugar. Ahora, las palabras más comunes son "darcy" (29 ocurrencias), "elizabeth" (25 ocurrencias), "bingley" (13 ocurrencias) y "bennet" (8 ocurrencias).

```python
wordcloud = WordCloud(width=800, height=400, background_color='white')
wordcloud.generate_from_text(' '.join(filtered_tokens_stopwords))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

La visualización de la nube de palabras (wordcloud) confirma de manera más impactante lo que hemos observado en la tabla anterior. Sin embargo, resulta interesante notar que la palabra "mr" aparece dos veces en nuestra representación visual.

**Stemming**

A continuación, empleamos el método de stemming, el cual se refiere a la reducción de palabras a su raíz o forma base. Esto implica eliminar sufijos y prefijos para obtener la parte central o significativa de una palabra.

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in filtered_tokens_stopwords]
stemmed_tokens
```

```python
freq_dist_stemmed = FreqDist(stemmed_tokens)

df_stemmed = pd.DataFrame([(word, freq) for word, freq in freq_dist_stemmed.items()], columns=['Palabra', 'Frecuencia'])

df_stemmed.sort_values(by='Frecuencia', ascending=False).reset_index(drop=True)
```

Al observar el proceso de eliminación de stopwords y aplicación de stemming, notamos que el número de palabras se redujo de 300 a 277. Esta reducción se debe a que, al aplicar el proceso de stemming, algunas palabras fueron reducidas a su forma base, lo que implica una pérdida parcial de su significado original. El stemming se basa en reglas heurísticas para eliminar sufijos y prefijos, sin considerar el contexto o el significado completo de las palabras.

Una observación interesante es que la palabra "mr" ahora tiene 40 ocurrencias. Recordemos que en la visualización de la nube de palabras con stopwords, notamos repeticiones de la palabra "mr". Al aplicar el stemming, estas repeticiones se han unificado en una sola forma, lo que explica el incremento en el recuento de ocurrencias de "mr".

```python
wordcloud = WordCloud(width=800, height=400, background_color='white')
wordcloud.generate_from_text(' '.join(stemmed_tokens))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Realiza análisis de sentimientos en tu conjunto de datos.

```python
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
scores = sia.polarity_scores(text)
df_scores = pd.DataFrame(scores, index=[0])
df_scores
```

```python
import seaborn as sns

sns.barplot(df_scores)
plt.xlabel ('Sentimientos')
plt.title ('Analisis Sentimientos')
plt.show()
```

Los resultados obtenidos mediante el uso de SentimentIntensityAnalyzer, que hemos observado tanto en la tabla como en el gráfico de barras, incluyen cuatro valores clave: 'neg', 'neu', 'pos' y 'compound'.

Cada uno de estos valores representa una métrica de sentimiento calculada para el texto del Resumen de Pride and Prejudice.

El valor 'neg' representa la proporción de negatividad en el texto. En este caso, el valor es 0.092, lo que indica que aproximadamente el 9.2% del texto tiene connotaciones negativas.

El valor 'neu' representa la proporción de neutralidad en el texto. En este caso, el valor es 0.773, lo que indica que aproximadamente el 77.3% del texto se considera neutral.

El valor 'pos' representa la proporción de positividad en el texto. En este caso, el valor es 0.135, lo que indica que aproximadamente el 13.5% del texto tiene connotaciones positivas.

El valor 'compound' representa el puntaje compuesto del sentimiento del texto en su conjunto. En este caso, el valor es 0.991, lo que indica un sentimiento muy positivo en el texto analizado.

En resumen, según los valores de esta tabla, el texto tiene una proporción baja de negatividad, una proporción alta de neutralidad y una proporción moderada de positividad. El puntaje compuesto indica que el texto tiene un sentimiento general muy positivo, lo que va de acuerdo con los aspectos fundamentales de la historia de Pride and Prejudice.
